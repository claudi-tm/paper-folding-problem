##Problem 1

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2**43
print("Espessura: {} metros".format(folded_thickness))
```

## Problem 3

```python

folded_thickness = THICKNESS * 2
for x in range(42):
  folded_thickness *= 2
print("Programming gives the same result with different paths.\nThickness:{} metros ".format(folded_thickness))
```

## Problem 4

```python
import time
start = time.time()
elapsed_time = time.time() - start

def powderCalculation():
  print("====================Calculation with powder===============")
  folded_thickness = THICKNESS * 2**43
  print("Thickness: {}[m]".format(folded_thickness))
  elapsed_time = time.time() - start
  print("time: {}[s]".format(elapsed_time))
  print("=============================================================")
powderCalculation()

def forCaculation():
  folded_thickness = THICKNESS * 2
  for x in range(42):
    folded_thickness *= 2
  print("====================Calculation with for loop===============")
  print("Thickness: {}[m]".format(folded_thickness))
  elapsed_time = time.time() - start
  print("time: {}[s]".format(elapsed_time))
  print("=============================================================")
forCaculation()
```

## Problem 5

```python
folded_thickness = THICKNESS * 2
thicknessList = []
for x in range(44):
  thicknessList.append("{}m".format(folded_thickness))
  folded_thickness *= 2
print("Thickness: {}[m]".format(thicknessList))
```

## Problem 6
The graphic desmotrates an increase of the values of thickness by the times the paper is folded

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=5)
plt.plot(thicknessList)
plt.show()

print(len(thicknessList))

%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=5, color="red")
plt.plot(thicknessList, color="red")
plt.show()

%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=5, color="green")
plt.plot(thicknessList, color="green")
plt.show()
```
